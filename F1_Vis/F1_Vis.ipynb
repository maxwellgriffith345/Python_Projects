```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set()
```

```python
data = pd.read_csv('Ham_Ver2020.csv', parse_dates = True)
```

```python
data.head()
```

```python
sns.lineplot(data = data, x="RaceDate", y="PTS", hue = 'DriverLast')
```

```python
data.sort_values(by = 'RaceDate')
```

```python
sns.lineplot(data = data, x="RaceDate", y="PTS", hue = 'DriverLast')
```
