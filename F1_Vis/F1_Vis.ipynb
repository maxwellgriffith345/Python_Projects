```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set()
```

```python
data = pd.read_csv('Ham_Ver2020.csv', parse_dates = ['RaceDate'])
```

```python
data.head()
```

```python
sns.lineplot(data = data, x="RaceDate", y="PTS", hue = 'DriverLast')
```

```python
data.sort_values(by = 'RaceDate')
```

```python
sns.lineplot(data = data, x="RaceDate", y="PTS", hue = 'DriverLast')
```

```python
data['TotalPTS'] = data.groupby(['DriverLast'])['PTS'].cumsum()
```

```python
sns.lineplot(data = data, x="RaceDate", y="TotalPTS", hue = 'DriverLast')
```

```python
graph=sns.lineplot(
    data = data, 
    x="RaceCountry", y="TotalPTS", hue = 'DriverLast')
graph.tick_params(axis = 'x', labelrotation = 90)
```

```python
graph=sns.lineplot(
    data = data, 
    x="RaceCountry", y="TotalPTS", hue = 'DriverLast',
    style = "DriverLast", markers = True, dashes = False)
graph.tick_params(axis = 'x', labelrotation = 90)
```

```python
data21 = pd.read_csv('Ham_Ver2021.csv', parse_dates = ['RaceDate'])
data21.head()
```

```python
Ham_Ver21 = data21[['DriverLast', 'PTS', 'RaceCountry', 'RaceDate']]
Ham_Ver21.head()
```

```python
Ham_Ver21.sort_values(by = 'RaceDate')
Ham_Ver21['TotalPTS'] = Ham_Ver21.groupby(['DriverLast'])['PTS'].cumsum()
Ham_Ver21.head()
```

```python
fig,axes = plt.subplots(1,2,figsize=(15,5), sharex=True)
fig.suptitle('2021 Drivers Champ Battle')

g1 = sns.lineplot(data = Ham_Ver21, x="RaceCountry", y="PTS", hue = 'DriverLast', ax=axes[0])
g1.tick_params(axis = 'x', labelrotation = 90)

g2 = sns.lineplot(data = Ham_Ver21, x="RaceCountry", y="TotalPTS", hue = 'DriverLast',ax = axes[1])
g2.tick_params(axis = 'x', labelrotation = 90)
```

```python
stacked_21 = pd.pivot(Ham_Ver21, index = 'RaceCountry', columns = ['DriverLast'], values = 'PTS')
stacked_21.head()
```

```python
stacked_21.plot(kind = 'bar', stacked = True)
```

```python
fig, axes = plt.subplots(figsize = (10,10))
graph =sns.barplot(data = Ham_Ver21, x = "RaceCountry", y = "PTS", hue = 'DriverLast')
graph.tick_params(axis = 'x', labelrotation = 90)
```

```python
wide_21 = pd.pivot(Ham_Ver21,index = ['RaceDate', 'RaceCountry'], columns = ['DriverLast'], values = 'TotalPTS')
wide_21['PTSDiff']= wide_21['Hamilton']-wide_21['Verstappen']
wide_21.head()
```

```python
wide_21.reset_index(inplace=True)
wide_21.head()
```

```python

g3 = sns.lineplot(data = wide_21, x = 'RaceCountry', y='PTSDiff')
g3.tick_params(axis = 'x', labelrotation = 90)
plt.ylim(-35,35)
#ax.set(ylim=(-35,35))
```

```python
fig, axes = plt.subplots()

g3 = sns.lineplot(data = wide_21, x = 'RaceCountry', y='PTSDiff', ax = axes, color = 'black')

g3.tick_params(axis = 'x', labelrotation = 90)
plt.ylim(-35,35)

axes.axhspan(0, 35, facecolor='blue', alpha=0.3)
axes.axhspan(-35, 0, facecolor='orange', alpha=0.3)
```

```python
fig, axes = plt.subplots(figsize = (10,10))

g3 = sns.lineplot(data = wide_21, x = 'RaceCountry', y='PTSDiff', ax = axes, color = 'black')

g3.tick_params(axis = 'x', labelrotation = 90)
plt.ylim(-35,35)

axes.axhspan(0, 35, facecolor='blue', alpha=0.3)
axes.axhspan(-35, 0, facecolor='orange', alpha=0.3)

axes.annotate('Ham Rams Ver', xy=('Great Britain', 10),
                 xytext = ('Great Britain', 10),
                 horizontalalignment = 'center')

axes.vlines('Great Britain', -35, 35, colors = 'r', linestyle = 'dashed')
             
```

```python
fig,axes = plt.subplots(2,2,figsize=(15,17), sharex=False)
fig.suptitle('2021 Drivers Champ Battle')

g1 = sns.barplot(data = Ham_Ver21, x = "RaceCountry", y = "PTS", hue = 'DriverLast', ax=axes[0,0])
g1.tick_params(axis = 'x', labelrotation = 90)

g2 = sns.lineplot(data = Ham_Ver21, x="RaceCountry", y="TotalPTS", hue = 'DriverLast',ax = axes[0,1])
g2.tick_params(axis = 'x', labelrotation = 90)

g3 = sns.lineplot(data = wide_21, x = 'RaceCountry', y='PTSDiff', ax = axes[1,0], color = 'black')

g3.tick_params(axis = 'x', labelrotation = 90)
axes[1,0].set_ylim([-35,35])

axes[1,0].axhspan(0, 35, facecolor='blue', alpha=0.3)
axes[1,0].axhspan(-35, 0, facecolor='orange', alpha=0.3)

axes[1,0].annotate('Ham Rams Ver', xy=('Great Britain', 10),
                 xytext = ('Great Britain', 10),
                 horizontalalignment = 'center')

axes[1,0].vlines('Great Britain', -35, 35, colors = 'r', linestyle = 'dashed')
```

```python
fig = plt.figure(figsize=(15,17), constrained_layout=True)
fig.suptitle('2021 Drivers Champ Battle')

ax1 = fig.add_subplot(2, 2, 1)
ax1=sns.barplot(data = Ham_Ver21, x = "RaceCountry", y = "PTS", hue = 'DriverLast')
ax1.tick_params(axis = 'x', labelrotation = 90)
ax1.set_title("Points Per Race")

ax2 = fig.add_subplot(2, 2, 2)
ax2=sns.lineplot(data = Ham_Ver21, x="RaceCountry", y="TotalPTS", hue = 'DriverLast')
ax2.tick_params(axis = 'x', labelrotation = 90)
ax2.set_title("Total Points")

ax3 = fig.add_subplot(2, 2, 3)
ax3=sns.lineplot(data = wide_21, x = 'RaceCountry', y='PTSDiff', color = 'black')
ax3.tick_params(axis = 'x', labelrotation = 90)
ax3.set_title("Points Gap")
ax3.set_ylim([-35,15])

ax3.axhspan(0, 15, facecolor='blue', alpha=0.3)
ax3.axhspan(-35, 0, facecolor='orange', alpha=0.3)

ax3.annotate('Ham Rams Ver', xy=('Great Britain', 10),
                 xytext = ('Great Britain', 10),
                 horizontalalignment = 'center')

ax3.vlines('Great Britain', -35, 15, colors = 'r', linestyle = 'dashed')


fig.savefig('Ham_Ver21.jpg')

```
