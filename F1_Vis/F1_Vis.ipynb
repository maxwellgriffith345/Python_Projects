```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set()
```

```python
data = pd.read_csv('Ham_Ver2020.csv', parse_dates = True)
```

```python
data.head()
```

```python
sns.lineplot(data = data, x="RaceDate", y="PTS", hue = 'DriverLast')
```

```python
data.sort_values(by = 'RaceDate')
```

```python
sns.lineplot(data = data, x="RaceDate", y="PTS", hue = 'DriverLast')
```

```python
data['TotalPTS'] = data.groupby(['DriverLast'])['PTS'].cumsum()
```

```python
data
```

```python
sns.lineplot(data = data, x="RaceDate", y="TotalPTS", hue = 'DriverLast')
```

```python
sns.lineplot(data = data, x="Circuit", y="TotalPTS", hue = 'DriverLast')
```

```python
graph=sns.lineplot(
    data = data, 
    x="RaceCountry", y="TotalPTS", hue = 'DriverLast')
graph.tick_params(axis = 'x', labelrotation = 90)
```

```python
graph=sns.lineplot(
    data = data, 
    x="RaceCountry", y="TotalPTS", hue = 'DriverLast',
    style = "DriverLast", markers = True, dashes = False)
graph.tick_params(axis = 'x', labelrotation = 90)
```
